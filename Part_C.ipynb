### What is Logistic Regression?

Despite the word **"Regression"** in its name, **Logistic Regression is a classification algorithm.** It is used to predict the probability of a categorical dependent variable.

1. **Classification, Not Regression**: While Linear Regression predicts continuous values (like a house price or an exam score), Logistic Regression predicts the probability that an input belongs to a specific category (like 'Pass' or 'Fail', 'Spam' or 'Not Spam').
2. **The Sigmoid Function**: It uses a mathematical function called the **Sigmoid function** (or Logistic function) to map any real-valued number into a value between 0 and 1.
3. **Decision Threshold**: By default, if the predicted probability is greater than 0.5, the model classifies the input into the positive class (e.g., 'Pass'); otherwise, it classifies it into the negative class (e.g., 'Fail').

#### Summary:
- **Linear Regression**: Predicts a number (e.g., 85.5%).
- **Logistic Regression**: Predicts a class (e.g., 'Pass').

### What is a Confusion Matrix?

A **Confusion Matrix** is a table used to evaluate the performance of a classification model. It compares the **Actual** values from the test data with the **Predicted** values generated by the model.

#### It represents four key metrics:

1.  **True Positives (TP)**: The model correctly predicted the positive class (e.g., predicted 'Pass' and the student actually 'Passed').
2.  **True Negatives (TN)**: The model correctly predicted the negative class (e.g., predicted 'Fail' and the student actually 'Failed').
3.  **False Positives (FP)**: The model incorrectly predicted the positive class (e.g., predicted 'Pass' but the student actually 'Failed'). Also known as a **Type I error**.
4.  **False Negatives (FN)**: The model incorrectly predicted the negative class (e.g., predicted 'Fail' but the student actually 'Passed'). Also known as a **Type II error**.

#### Why is it useful?
Instead of just looking at overall **Accuracy**, the confusion matrix helps you see if your model is biased or struggling with a specific class (for example, if it's failing to identify students who are actually at risk of failing).
